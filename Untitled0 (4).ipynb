```python
START_YEAR = 2009

# 2. Моделируем «ожидаемую цену» на основе тональности
# ───────────────────────────────────────────────────────────────
expected_price = []

for i, (_, row) in enumerate(merged.iterrows()):
    if i == 0:
        expected_price.append(row['Close'])
        continue

    prev_price = expected_price[-1]

    if (row['Date'].year < START_YEAR) or pd.isna(row['avg_sentiment']):
        expected_price.append(prev_price)
    elif row['avg_sentiment'] > 0:
        expected_price.append(prev_price * 1.07)   # +7 %
    elif row['avg_sentiment'] < 0:
        expected_price.append(prev_price * 0.93)   # −7 %
    else:
        expected_price.append(prev_price)

merged['expected_price'] = expected_price

# ───────────────────────────────────────────────────────────────
# 3. Для графика оставляем только даты ≥ 2009
# ───────────────────────────────────────────────────────────────
plot_df = merged[merged['Date'].dt.year >= START_YEAR]

plt.figure(figsize=(14, 6))
plt.plot(plot_df['Date'], plot_df['Close'], label='Реальная цена')
plt.plot(plot_df['Date'], plot_df['expected_price'],
         label='Ожидаемая по тональности', linestyle='--')
plt.title('Honda: real vs.predicted)')
plt.xlabel('Дата')
plt.ylabel('Цена акции')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
import numpy as np

def evaluate_knn_baseline(data, features, target='Close', test_size=300, k=5):
    """
    Построить KNN-модель и вернуть метрики RMSE, MAPE, R2.
    """
    df = data.dropna(subset=features + [target]).copy()

    X = df[features]
    y = df[target]

    X_train, X_test = X[:-test_size], X[-test_size:]
    y_train, y_test = y[:-test_size], y[-test_size:]

    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))  # вручную
    mape = mean_absolute_percentage_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    return {'model': model, 'RMSE': rmse, 'MAPE': mape, 'R2': r2, 'y_true': y_test, 'y_pred': y_pred}
```

```python
# -------------------------------------------------------------
#  A. ОБЩИЙ КЛАСС FinBERT-регрессора и токенизатор
# -------------------------------------------------------------
tok  = AutoTokenizer.from_pretrained("ProsusAI/finbert")
bert = AutoModel.from_pretrained("ProsusAI/finbert")

class FinBERTReg(nn.Module):
    def __init__(self, bert, macro=False):
        super().__init__()
        self.bert = bert
        self.regressor = nn.Linear(768 + int(macro), 1)
    def forward(self, ids, mask, sp500=None):
        cls = self.bert(ids, attention_mask=mask).last_hidden_state[:, 0]
        if sp500 is not None:
            if sp500.dim() == 1: sp500 = sp500.unsqueeze(1)
            cls = torch.cat((cls, sp500), 1)
        return self.regressor(cls).squeeze()


# -------------------------------------------------------------
#  B. Датасет для «чистых» моделей  (без SP500)
# -------------------------------------------------------------
class NewsDS(Dataset):
    def __init__(self, texts, targets):
        self.texts, self.targets = texts, targets
    def __len__(self):  return len(self.texts)
    def __getitem__(self, i):
        enc = tok(self.texts[i], padding='max_length',
                  truncation=True, max_length=128, return_tensors='pt')
        return {'ids':enc['input_ids'].squeeze(),
                'mask':enc['attention_mask'].squeeze(),
                'y':torch.tensor(self.targets[i],dtype=torch.float32)}

# -------------------------------------------------------------
#  C. Датасет для моделей  с SP500
# -------------------------------------------------------------
class NewsMacroDS(NewsDS):
    def __init__(self, texts, targets, sp500):        #  ⟵ добавили sp500
        super().__init__(texts, targets); self.sp500=sp500
    def __getitem__(self, i):
        item = super().__getitem__(i)
        item['sp500'] = torch.tensor(self.sp500[i], dtype=torch.float32)
        return item
```

```python
def infer_model(model, loader, use_macro=False):
    model.eval(); preds, trues = [], []
    with torch.no_grad():
        for batch in loader:
            ids  = batch['ids' ].to(device)
            mask = batch['mask'].to(device)
            y    = batch['y'   ].to(device)
            if use_macro:
                sp  = batch['sp500'].to(device)
                out = model(ids, mask, sp)
            else:
                out = model(ids, mask)
            preds.extend(out.cpu().tolist())
            trues.extend(y.cpu().tolist())
    return np.array(trues), np.array(preds)
```

```python
# -------------------------------------------------------------
#  IMPACT_SCORE для HMC, INTC, GOOGL
# -------------------------------------------------------------
import pandas as pd, os, warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

PRICE_DIR = "/content/drive/MyDrive/Colab Notebooks"

price_spec = {
    'HMC'  : {'file': "Honda_Data.csv",  'close_col': 'Close'},
    'GOOGL': {'file': "Google 2020-2025 Stock Market.csv", 'close_col': 'Close'},
    'INTC' : {'file': "intel_data.csv", 'close_col': 'Close'}
}

impact_frames = []

for sym, spec in price_spec.items():
    path       = os.path.join(PRICE_DIR, spec['file'])
    close_col  = spec['close_col']

    # --- читаем только Date и Close -----------------------------------------
    px = pd.read_csv(path, usecols=lambda c: c in ['Date', close_col])

    # --- переводим Date в tz-naive YYYY-MM-DD -------------------------------
    px['Date'] = (
        pd.to_datetime(px['Date'], errors='coerce', utc=True)  # mixed → UTC
          .dt.tz_localize(None)                               # убираем пояс
          .dt.normalize()                                     # 00:00:00
    )
    px = px.dropna(subset=['Date'])                           # выбрасываем битые даты

    # --- impact_score = |Close_{t+1} − Close_t| / Close_t -------------------
    px = px.sort_values('Date').reset_index(drop=True)
    px['impact_score'] = (px[close_col].shift(-1) - px[close_col]).abs() / px[close_col]
    px = px.dropna(subset=['impact_score'])                   # последняя строка NaN

    impact_frames.append(
        px[['Date', 'impact_score']].assign(Stock_symbol=sym)
    )

impact_all = pd.concat(impact_frames, ignore_index=True)

print("✅ impact_all готов:", impact_all.shape)
display(impact_all.head())
```

```python
# ----------------- imports & helpers --------------------------------
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
import yfinance as yf, datetime as dt, os, re, warnings, torch, torch.nn as nn
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, f1_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel
warnings.filterwarnings("ignore")
plt.style.use('seaborn-v0_8-whitegrid')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def rmse(a,b):  return np.sqrt(mean_squared_error(a,b))
def mape(a,b):  return mean_absolute_percentage_error(a,b)
def f1_dir(a,b):
    a_dir = np.sign(np.diff(a, prepend=a[0]))
    b_dir = np.sign(np.diff(b, prepend=b[0]))
    return f1_score(a_dir, b_dir, average='macro')
```

```python
# ----------------- 2.1 цены трёх тикеров → impact_score ------------
PRICE_DIR = "/content/drive/MyDrive/Colab Notebooks"
price_spec = {
    'HMC'  : ("Honda_Data.csv"                       , 'Close'),
    'INTC' : ("intel_data.csv"                       , 'Close'),
    'GOOGL': ("Google 2020-2025 Stock Market.csv"    , 'Close')
}
impact_frames=[]
for sym,(file,close_col) in price_spec.items():
    df = pd.read_csv(os.path.join(PRICE_DIR,file),
                     usecols=lambda c: c in ['Date',close_col])
    df['Date'] = (pd.to_datetime(df['Date'], utc=True)
                    .dt.tz_localize(None).dt.normalize())
    df  = df.sort_values('Date').reset_index(drop=True)
    df['impact_score'] = (df[close_col].shift(-1)-df[close_col]).abs()/df[close_col]
    df  = df.dropna(subset=['impact_score'])
    impact_frames.append(df[['Date','impact_score']]
                         .assign(Stock_symbol=sym))
impact_all = pd.concat(impact_frames, ignore_index=True)

# ----------------- 2.2 новости + S&P-500 ---------------------------
NEWS_CSV = "/content/drive/MyDrive/Colab Notebooks/filtered_news.csv"
news = pd.read_csv(NEWS_CSV).rename(columns={'tock_symbol':'Stock_symbol'})
news['Date'] = (pd.to_datetime(news['Date'], errors='coerce', utc=True)
                  .dt.tz_localize(None).dt.normalize())

sp = yf.download("^GSPC", start="2009-08-22",
                 end=dt.date.today().isoformat(),
                 auto_adjust=False, progress=False)
sp.columns = ["_".join(c).strip() if isinstance(c,tuple) else c for c in sp.columns]
close_col = next(c for c in sp.columns if re.search(r'close',c,re.I))
sp = (sp[[close_col]].rename(columns={close_col:'SP500_Close'})
        .reset_index())
sp['Date'] = (pd.to_datetime(sp['Date'], utc=True)
                .dt.tz_localize(None).dt.normalize())
sp['SP500_Return'] = sp['SP500_Close'].pct_change()

df_all = (news
          .merge(impact_all, on=['Stock_symbol','Date'], how='left')
          .merge(sp[['Date','SP500_Return']], on='Date', how='left')
          .sort_values('Date').reset_index(drop=True))
df_all['impact_score'].fillna(0.0, inplace=True)
df_all['SP500_Return'] = df_all['SP500_Return'].fillna(method='ffill').fillna(0.0)
print("✅ df_all shape:", df_all.shape)
```

```python
# ----------------- FinBERT модель-обёртка --------------------------
tok  = AutoTokenizer.from_pretrained("ProsusAI/finbert")
bert = AutoModel.from_pretrained("ProsusAI/finbert")

class FinBERTReg(nn.Module):
    def __init__(self, bert, macro=False):
        super().__init__()
        self.bert = bert
        self.regressor = nn.Linear(768+int(macro),1)
    def forward(self, ids, mask, macro=None):
        cls = self.bert(ids, attention_mask=mask).last_hidden_state[:,0]
        if macro is not None:
            if macro.dim()==1: macro=macro.unsqueeze(1)
            cls = torch.cat((cls, macro),1)
        return self.regressor(cls).squeeze()

class NewsDS(Dataset):
    def __init__(self, texts, y):
        self.texts, self.y = texts, y
    def __len__(self): return len(self.texts)
    def __getitem__(self,i):
        enc = tok(self.texts[i], padding='max_length',
                  truncation=True, max_length=128, return_tensors='pt')
        return {'ids':enc['input_ids'].squeeze(),
                'mask':enc['attention_mask'].squeeze(),
                'y':torch.tensor(self.y[i],dtype=torch.float32)}

class NewsMacroDS(NewsDS):
    def __init__(self,texts,y,macro): super().__init__(texts,y); self.mac=macro
    def __getitem__(self,i):
        d=super().__getitem__(i); d['macro']=torch.tensor(self.mac[i],dtype=torch.float32); return d

def infer(model, loader, use_macro=False):
    model.eval(); pr, tr = [],[]
    with torch.no_grad():
        for b in loader:
            ids=b['ids'].to(device); mask=b['mask'].to(device); y=b['y'].to(device)
            out = model(ids, mask, b['macro'].to(device) if use_macro else None)
            pr.extend(out.cpu().tolist()); tr.extend(y.cpu().tolist())
    return np.array(tr), np.array(pr)
```

```python
# ============================ METRICS: PRICE PREDICTION (Intel Close) ==============================

# Проверим, что y_true_price уже есть (300 последних значений Intel Close)
print("ℹ️ len(y_true_price):", len(y_true_price))

# Словарь моделей
models_price = {
    'KNN'  : y_pred_knn,
    'LSTM' : y_pred_lstm,
    'GRU'  : y_pred_gru,
}

# Надёжная MAPE
def robust_mape(y_true, y_pred, eps=1e-3):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    mask = np.abs(y_true) > eps
    if np.sum(mask) == 0:
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask]))

# Сбор метрик
rows = []
for name, pred in models_price.items():
    if isinstance(pred, (list, np.ndarray)) and len(pred) == len(y_true_price):
        rows.append({
            'Model': name,
            'RMSE' : rmse(y_true_price, pred),
            'MAPE' : robust_mape(y_true_price, pred),
            'F1'   : f1_dir(y_true_price, pred),
        })
    else:
        print(f"⚠️ {name} пропущен: len(pred) = {len(pred)}, len(y_true_price) = {len(y_true_price)}")

# Вывод таблицы
if rows:
    df_price = pd.DataFrame(rows).set_index('Model').sort_values('RMSE')
    display(df_price.style.format({'RMSE': '{:.3f}', 'MAPE': '{:.2%}', 'F1': '{:.3f}'}))
else:
    print("❌ Нет подходящих моделей для Intel Close.")
```
